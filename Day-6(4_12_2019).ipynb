```
# while loop
# print numbers from 1 to N
n = int(input("enter a number"))
i = 1
while i <= n:
  print(i, end="  ")
  i = i + 1
```

```
# print all even natural numbers from 1to N
n = int(input("enter a number"))
i = 1
while i<=n:
  if i%2==0:
    print(i, end="  ")
  i=i+1
```

```
#  print sum of even numbers from 1 to N
n=int(input("enter a number"))
i=1
sum=0
while i<=n:
  if i%2==0:
    sum=sum+i
  i=i+1
print(sum)
```

```
# extracting the digits from the number
n=int(input("enter the number"))
while n!=0:
  print(n%10, end="  ")
  n=n//10
```

# whether the given number is palindrome or not
# whether the given number is armstrong series

```
n=int(input("enter the number"))
rev=0
temp=n
while n!=0:
  r=n%10
  rev=rev*10+r
  n=n//10
if rev==temp:
  print("palindrome")
else:
  print("not a palindrome")
```
