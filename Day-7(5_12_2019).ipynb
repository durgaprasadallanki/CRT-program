```
def even_sum(n):
  sum=0
  for i in range(1,n+1):
    if i%2==0:
      sum+=i
  return sum
print(even_sum(10))
print(even_sum(30))
```

```
# factorial of a number using functions
def factorial(n):
  f=1
  for i in range(1,n+1):
    f=f*i
  print(f)
(factorial(5))
(factorial(10))
```
