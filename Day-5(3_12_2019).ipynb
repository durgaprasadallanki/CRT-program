## operators

1.   arithmetic operators------>+,-,*,/,%
2.   logical operators------>and,or,not
3.   relational operator----->==,!=,<,>,<=,>=
4.   membership operator----->in,not in



```
# arithmatic operation
a = 12
b = 23
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
print(a//b)
```

```
# logical operator
i = 10
j = 20
print(i>1 and i<20)
print(i>12 or i<33)
print(not i)
```

```
# relational operator
print(i==j)
print(i!=j)
print(i>j)
print(i<j)
print(i>=j)
print(i<=j)
```

```
# membership operator
str1 = "gitam"
print('z' in str1)
print('z' not in str1)
```
